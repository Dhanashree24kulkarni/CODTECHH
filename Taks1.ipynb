```python
# Cell 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cell 2: Load the dataset
from sklearn.datasets import load_iris

# Load Iris dataset and convert it to a Pandas DataFrame
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})  # Map target to species names
df.head()  # Display the first few rows of the dataset
```

```python
# Cell 3: Dataset information and statistics
print("Dataset Info:")
df.info()  # Overview of columns, types, and non-null counts

print("\nSummary Statistics:")
print(df.describe())  # Summary statistics for numerical columns
```

```python
# Cell 4: Check for missing values and duplicates
print("Missing Values in Each Column:")
print(df.isnull().sum())  # Count missing values

print("\nNumber of Duplicate Rows:")
print(df.duplicated().sum())  # Count duplicate rows
```

```python
# Cell 5: Histograms for all numerical columns
df.hist(figsize=(12, 10), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Feature Distributions')
plt.show()
```

```python
# Cell 6: Scatter plot example (Sepal Length vs Sepal Width)
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=df)
plt.title('Sepal Length vs Sepal Width by Species')
plt.show()
```

```python
# Cell 7: Correlation heatmap
plt.figure(figsize=(10, 8))
# Select only numerical columns for correlation
correlation_matrix = df.select_dtypes(include=[np.number]).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Cell 8: Box plots for each numerical feature
for col in df.columns[:-1]:  # Exclude the species column
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df[col], color='lightblue')
    plt.title(f'Box Plot of {col}')
    plt.show()
```

```python
# Cell 9: Save the dataset as a CSV file
df.to_csv('iris_dataset_cleaned.csv', index=False)
print("Cleaned dataset saved as 'iris_dataset_cleaned.csv'.")
```
